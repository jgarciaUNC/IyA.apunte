# ADF diagrama radiación

```python
#-----------------------------Constantes
reset()

Pi = pi.n()

# para elevar el número "e" a una potencia hacer "exp()"

# la unidad imaginaria en SageTex es "I"

#-----------------------------Funciones auxiliares

def Numerico(datos):
    """
    Convierte una lista con expresiones en sus valores numericos
    """
    return [datos[i].n() for i in [0..len(datos) - 1 ]]   

import csv        # permite importar y exportar archivos .csv

```

```python
Origen = 0.  + I * 0.     # Origen figuras

N = 100                   # Discretización

Rcirculo = 1. 

theta = [( k * 2 * Pi / N )  for k in [0..N] ]  # [rad] ángulo discretización círculo

Circulo = [( Rcirculo * exp( I * theta[k] ))  for k in [0..N] ]  # Circulo discretizado plano Z

Omni = list_plot( Circulo , aspect_ratio = 1 , \
             # axes_labels = ['$xi$' , '$eta$' ] , \
             # title_pos=( 1. , 1. ) , \
             # title = 'Antena Omnidireccional' ,  \
                 plotjoined=True , color = 'red')

Omni
```

```python
CuadroPosit = [( Rcirculo  * cos(theta[k]) * exp( I * theta[k] ))  for k in [0..N] ]  # Circulo discretizado plano Z
CuadroNegat = [(- Rcirculo * cos(theta[k]) * exp( I * theta[k] ))  for k in [0..N] ]  # Circulo discretizado plano Z

OO1 = list_plot( CuadroPosit , aspect_ratio = 1 , \
             # axes_labels = ['$xi$' , '$eta$' ] , \
             # title_pos=( 1. , 1. ) , \
             # title = 'Antena Omnidireccional' ,  \
                 plotjoined=True , color = 'blue')

OO2 = list_plot( CuadroNegat , aspect_ratio = 1 , \
             # axes_labels = ['$xi$' , '$eta$' ] , \
             # title_pos=( 1. , 1. ) , \
             # title = 'Antena Omnidireccional' ,  \
                 plotjoined=True , color = 'green')

OO1 + OO2 + Omni
```

```python
Cardioide = [( Circulo[k] + CuadroPosit[k]  )  for k in [0..N] ]  # Circulo discretizado plano Z



Card = list_plot( Cardioide , aspect_ratio = 1 , \
                 plotjoined=True , color = 'pink')

Card + Omni + OO1 
```
